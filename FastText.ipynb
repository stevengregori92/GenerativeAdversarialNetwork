```python
import os
import pandas as pd

from tqdm.auto import tqdm
from nltk.tokenize import word_tokenize
from gensim.models import FastText
```

#Prepare Corpus

```python
!gdown https://drive.google.com/uc?id=10OnDq1OAByUxcQEPCDZ19jxU0W8QV6hu
!unzip /content/word2vec.zip
```

```python
df = pd.read_csv('data/gojek_review.csv')
df
```

```python
sentences = [word_tokenize(review.lower()) for review in tqdm(df.review)]
sentences[:5]
```

#Train FastText Model

```python
model = FastText(sentences, vector_size=128, window=5, min_count=3, workers=4, epochs=1000, sg=0, hs=0)
```

#Save

```python
os.makedirs("model/fasttext/", exist_ok=True)
```

```python
model.save('model/fasttext/gojek_review.fasttext')
```

#Load

```python
model = FastText.load('model/fasttext/gojek_review.fasttext')
```

#Continue Training

```python
contoh_data = [
    ['aplikasinya', 'ok', 'kok', '.', 'sangat', 'membantu', 'untuk', 'keseharian', 'saya'],
    ['aplikasinya', 'gak', 'nyaman', 'dipakai', 'bule']
]
```

```python
model.train(contoh_data, total_examples = len(contoh_data), epochs=1)
```

```python
model.save('model/fasttext/gojek_review.fasttext')
```

higher order visualization

```python
from umap import UMAP
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
X = UMAP().fit_transform(fasttext.vectors)
```

```python
df = pd.DataFrame(X, columns = ['umap1', 'umap2'])
df['text'] = fasttext.index_to_key
```

```python
df
```

```python
fig = px.scatter(df, x='umap1', y='umap2', text='text')
fig.update_traces(textposition='top center')
fig.update_layout(
    height = 800,
    title_text = 'Reduced FastText Visualization'
)
fig.show()
```
